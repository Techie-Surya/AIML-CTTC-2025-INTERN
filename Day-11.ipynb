##### to remove the unnecessary space present at the starting part of the string in each columns, we need to remove that.

```python
adult = pd.read_csv(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\adult\adult.data",header = None,delimiter=' *, *')
adult
```

```python
adult.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
adult
```

```python
for i in adult.columns:
    print(f"{i}: \n {adult[i].unique()}\n")
```

```python
des=adult.describe(include='all')
des
```

```python
adult['workclass'].replace('?',des['workclass'][2],inplace=True)
```

```python
adult['occupation'].replace('?',des['occupation'][2],inplace=True)
```

```python
adult['native-country'].replace('?',des['native-country'][2],inplace=True)
```
