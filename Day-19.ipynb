```python
x = np.array([3,7,5,9,11,15])
y = np.array([5,9,2,6,15,17])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# Single Layer Perceptron:
model = Sequential()
# input & hidden layer:
model.add(Dense(1,input_dim = 1,activation = 'linear'))
# output layer:
model.add(Dense(1,activation = 'linear'))
model.compile(Adam(learning_rate = 0.01),loss = 'mse')
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=20,verbose=1)
```

```python
model.predict(x)
```

```python
ip = adv.drop('Sales',axis=1)
ip.head()
```

```python
op = adv.Sales
op.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(ip,op,train_size=0.8)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
# Applying Multilayer Perceptron Model
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# Creating the  Model
model = Sequential()
# input layer
model.add(Dense(30,input_dim=3, activation = 'linear'))
# hidden layer
model.add(Dense(15,activation = 'linear'))
model.add(Dense(25,activation = 'linear'))
# output layer
model.add(Dense(1,activation = 'linear'))
model.compile(Adam(learning_rate=0.01),loss = 'mse')
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs = 25)
```

```python
pred = model.predict(xtest)
pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(pred,ytest)
rs = r2_score(pred,ytest)
print(f"Mean Squarred Error: {mse} \nR2_score: {rs}")
```
