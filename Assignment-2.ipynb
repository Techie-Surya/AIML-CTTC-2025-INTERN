```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Auto Mpg Set

```python
auto = pd.read_csv(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\auto-mpg.csv")
auto
```

```python
auto.isnull().sum()
```

```python
auto.dtypes
```

```python
for i in auto.columns:
    print(f"{i}: \n {auto[i].unique()} \n")
```

```python
for i in auto.columns:
    print(f"{i}: {sum(auto[i]=='?')}")
```

```python
des=auto.describe(include='all')
des
```

```python
auto['horsepower'].replace('?',des['horsepower'][2],inplace=True)
```

```python
for i in auto.columns:
    print(f"{i}: {sum(auto[i]=='?')}")
```

```python
auto.horsepower = auto.horsepower.astype('float')
auto.dtypes
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
sns.distplot(auto.horsepower,color='green')
plt.show
```

```python
sns.displot(auto.horsepower,fill=False,color='red',hatch='--')
plt.show()
```

```python
sns.kdeplot(auto.horsepower,fill=True,hatch='*',color='yellow')
plt.show()
```

```python
sns.boxplot(x = auto.horsepower,y = auto.displacement,palette='cool')
plt.show()
```

```python
sns.violinplot(x = auto.horsepower,y = auto.displacement,palette='cool')
plt.show()
```

```python
sns.swarmplot(x = auto.horsepower,y = auto.displacement,palette='cool')
plt.show()
```

```python
sns.barplot(x = auto.horsepower,y = auto.displacement,palette='hot')
plt.show()
```

```python
plt.figure(figsize=(30,25))
sns.countplot(x = auto.horsepower,palette='deep')
plt.show()
```

```python
sns.scatterplot(x = auto.horsepower , y = auto.displacement , hue = auto.cylinders,palette='deep')
plt.show()
```

```python
sns.jointplot(data=auto)
plt.show()
```

```python
sns.jointplot(data=auto,kind='kde')
plt.show()
```

```python
sns.pairplot(auto)
plt.show()
```

```python
sns.pairplot(auto,hue = 'cylinders')
plt.show()
```
