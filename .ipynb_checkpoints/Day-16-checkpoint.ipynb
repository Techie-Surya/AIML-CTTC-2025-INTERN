# Unsupervised Machine Learning:-
<pre>Machine tries to group the datas based on some similar patterns or features</pre>
<ul><li>It refers to the algorithm that learns patterns from unlabelled data.</li></ul>

### Types of Unsupervised ML:-
* 1.Exclusive Unsuprevised ML
* 2.Overlapping Unsupervised ML
* 3.Hierarchical Unsupervised ML

### K-Means Clustering:- It belongs to exclusive unsupervised machine learning algorithm
<pre>Clustering :- Means grouping up of the datas</pre>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
iris = pd.read_csv(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\iris\iris.data",header=None)
iris
```

```python
iris.columns=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
iris
```

```python
iris.isnull().sum()
```

```python
iris.dtypes
```

```python
for i in iris.columns:
    print(f"{i}:\n {iris[i].unique()}\n")
```

```python
# We need to drop the column 'Class'
ip = iris.drop('Class',axis=1)
ip.head()
```

```python
# To apply the K means clustering Algorithm:-
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km.fit(ip)
```

```python
# Prediction
km_pred = km.predict(ip)
km_pred
```

```python
centroids = km.cluster_centers_
centroids
```

```python
iris['Predictions'] = km_pred
iris
```

```python
sns.scatterplot(x=iris.Sepal_Length,y=iris.Petal_Length)
plt.show()
```

```python
sns.scatterplot(x=iris.Sepal_Length,y=iris.Petal_Length,c = km_pred)
plt.show()
```

```python
# Plotting along the centroid values 
sns.scatterplot(x=iris.Sepal_Length,y=iris.Petal_Length,c = km_pred)
sns.scatterplot(x=centroids[:,0],y=centroids[:,2],s = 200 , marker = '*',color='k')
plt.show()
```
