### Iris Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
iris=pd.read_csv(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\iris\iris.data",header=None)
iris
```

```python
iris.columns=['SL','SW','PL','PW','CLASS']
iris
```

```python
iris.isnull().sum()
```

```python
iris.dtypes
```

```python
for i in iris.columns:
  print(f"{i}:\n{iris[i].unique()}")
```

```python
iris.describe(include='all')
```

```python
iris.info()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
iris.CLASS=le.fit_transform(iris.CLASS)
iris.CLASS.unique()
```

```python
le.inverse_transform([0, 1, 2])
```

```python
ip=iris.drop('CLASS',axis=1)
ip.head()
```

```python
op=iris.CLASS
op.head()
```

```python
# To convert the output columns to categorical:
from keras.utils import to_categorical
op = to_categorical(op)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(ip,op,train_size=0.8)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
xtest =  sc.fit_transform(xtest)
```

```python
# Model Training
model = Sequential()
#input layer 
model.add(Dense(20,input_dim = 4,activation = 'relu'))
#hidden layer
model.add(Dense(25,activation = 'relu'))
model.add(Dense(30,activation = 'relu'))
# output layer
model.add(Dense(3,activation = 'softmax'))
model.compile(Adam(learning_rate = 0.01),loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,validation_data = (xtest,ytest),epochs = 20)
```

```python
pred = model.predict(xtest)
pred
```

```python
p = np.argmax(pred, axis=1)
p
```

```python
ytest
```

```python
ac = np.argmax(ytest,axis = 1)
ac
```

```python
from sklearn.metrics import accuracy_score, recall_score
acc = accuracy_score(p,ac)
rs = recall_score(p,ac,average='macro')
print(f"Accuracy Score: {acc}")
print(f"Recall Score: {rs}")
```

### MNIST Dataset:- Using ANN model:-

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.datasets import mnist
```

```python
(xtrain,ytrain),(xtest,ytest) = mnist.load_data()
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
for i in range(0,10):
    plt.imshow(xtrain[ytrain==i][3000],cmap = 'gray')
    plt.title(str(i))
    plt.show()
```

```python
# to_categorical
from keras.utils import to_categorical
ytrain = to_categorical(ytrain,10)
ytest = to_categorical(ytest)
```

```python
xtrain = xtrain/255
xtest = xtest/255
```

```python
xtrain
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain = xtrain.reshape(60000,784)
xtest = xtest.reshape(10000,784)
```

```python
xtrain.shape
```

```python
# Model Training:
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
#input layer
model.add(Dense(35,input_dim = 784,activation =  'relu'))
#hidden layer
model.add(Dense(15,activation = 'relu'))
model.add(Dense(25,activation = 'relu'))
#output layer
model.add(Dense(10,activation = 'softmax'))
model.compile(Adam(learning_rate=0.01),loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
h = model.fit(xtrain,ytrain,validation_split = 0.1,epochs=20)
```

```python
plt.plot((h.history['loss']))
plt.plot((h.history['val_loss']))
plt.legend('loss','val_loss')
plt.title('loss')
plt.show()
```

```python
plt.plot((h.history['accuracy']))
plt.plot((h.history['val_accuracy']))
plt.legend('accuracy','val_accuracy')
plt.title('accuracy')
plt.show()
```
