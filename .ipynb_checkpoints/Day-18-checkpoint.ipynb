```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Face Detection Manually
face_detect = cv2.imread(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\RohitSharmaFace.webp")
plt.imshow(face_detect[::,::,::-1])
plt.grid()
plt.show()
```

```python
cv2.rectangle(face_detect,(150,50),(300,200),(255,0,0),4)
plt.imshow(face_detect[::,::,::-1])
plt.show()
```

```python
# Face detection using HaarCascade Classifier Algorithm:
face = r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\Haarcascade_frontalface_default.xml"
model = cv2.CascadeClassifier(face)
img = cv2.imread(r"C:\Users\surya\OneDrive\Desktop\AIML_CTTC\NOTES\RohitSharmaFace.webp")
facepoint = model.detectMultiScale(img)
for x,y,w,h in facepoint:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
plt.imshow(img[::,::,::-1])
plt.show()
```
