### Auto_Mpg dataset: 

```python
import numpy as np
import pandas as pd 
```

```python
# to check the datatype
auto.dtypes
```

```python
# to find the unique values
for i in auto.columns:
     print(f"{i}: \n {auto[i].unique()}\n")
```

```python
# to find the total number of missing values (?) in the dataset.
sum(auto['mpg'] == '?')
```

```python
for i in auto.columns:
    print(f"{i} : {sum(auto[i]=='?')}")
```

```python
des = auto.describe(include='all')
des
```

```python
# replacing the '?' with top repeated value
auto['horsepower'].replace('?',des['horsepower'][2])
```

```python
for i in auto.columns:
    print(f"{i} : {sum(auto[i]=='?')}")
```

```python
# to make the changes permanently 
auto['horsepower'].replace('?',des['horsepower'][2],inplace=True)
```

```python
for i in auto.columns:
    print(f"{i} : {sum(auto[i]=='?')}")
```

```python
auto.dtypes
```

```python
# to change the ddatatype of 'horsepower' column explicitly
auto.horsepower = auto.horsepower.astype('float')
auto.dtypes
```

```python
auto.describe(include='all')
```

## Matplotlib : It is Data Visualisation Tool.
<pre> Used to plot different graphs </pre>

```python
# Import the libraries:-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line Plot : is used to visualize a particular trend oveer time.

```python
x = [10,20,30,40,50,60]
y = [2000,2001,2002,2003,2004,2005]
plt.plot(x,y) # to the graph using x & y data
plt.show() # to plot the graph
```

```python
x = [45,89,23,67,45,67,27]
y = [2000,2001,2002,2003,2004,2005,2006]
plt.title('Line plot',c='red') # gives a heading to the graph
plt.plot(x,y,marker='o',linestyle='--',c='green')
plt.xlabel('Datas') # Gives a name to x-axis
plt.ylabel('Year') # gives a name to y-axis
plt.show()
```

## Bar plot :- used for comparison of categories.

```python
x = ['Team A','Team B','Team C']
y = [288,567,345]
colours = ['yellow','green','orange']
plt.bar(x,y,color = colours)  # plot a bar graph
plt.title('Bar Plot')
plt.xlabel('Teams')
plt.ylabel('No. of Members')
plt.show()
```
