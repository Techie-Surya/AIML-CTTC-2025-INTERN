```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Scatter Plot:- Used to display the relationship between two or more variables.

```python
x = [3,67,35,23,45,23,78,56,22]
y = [1,5,7,10,22,15,40,23,54]
plt.scatter(x,y,color='cyan',marker='s',s=200)
plt.title('Scatter Plot')
plt.show()
```

## Histogram:- Used to display the frequency of the data distribution.

```python
age = [15,25,65,52,47,86,31,26,75,7341,58]
bins = [10,20,30,40,50,60,70,80,90]
plt.hist(age,bins,color='gold',edgecolor='black',histtype='bar',rwidth=0.8)
plt.title('Histogram')
plt.xlabel('Interval')
plt.ylabel('Ages')
plt.show()
```

## Pie Plot/Pie Chart:- Used to display data as a proportion of any whole.

```python
activities = ['playing','sleeping','eating','roaming','studying']
slices = [6,12,2,3,1]
colours = ['cyan','gold','lime','pink','coral']
plt.pie(slices,labels=activities,colors=colours,startangle=180,shadow=True,explode=(0.1,0,0,0,0),autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Importing the libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Visualization:-
#### Distribution Plot:- Used to show the distribution of any single variable.
<pre>combines the histogram and kernal density estimate plot(KDE)</pre>

```python
# to remove the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Visualization:-
# Distribution Plot:-
sns.distplot(iris.Sepal_Length)
plt.show
```

##### Displot:- Shows single value histogram

```python
sns.displot(iris.Sepal_Length,fill=False,hatch='*',color='pink')
plt.show()
```

##### KDE Plot:- Used to display the frequency distribution of any single variable using curved lines 

```python
sns.kdeplot(iris.Sepal_Width,fill=True,hatch='o',color='cyan')
plt.show()
```

```python
sns.distplot(iris.Petal_Length[iris.Class=='Iris-setosa'],label='Setosa')
sns.distplot(iris.Petal_Length[iris.Class=='Iris-versicolor'],label='Versicolor')
sns.distplot(iris.Petal_Length[iris.Class=='Iris-virginica'],label='Virginica')
plt.legend()
plt.show
```

```python
sns.histplot(iris.Petal_Length[iris.Class=='Iris-setosa'],label='Setosa')
sns.histplot(iris.Petal_Length[iris.Class=='Iris-versicolor'],label='Versicolor')
sns.histplot(iris.Petal_Length[iris.Class=='Iris-virginica'],label='Virginica')
plt.legend()
plt.show
```

```python
sns.displot(iris.Petal_Length[iris.Class=='Iris-setosa'],label='Setosa',color='red')
sns.displot(iris.Petal_Length[iris.Class=='Iris-versicolor'],label='Versicolor',color='gold')
sns.displot(iris.Petal_Length[iris.Class=='Iris-virginica'],label='Virginica')
plt.legend()
plt.show
```

```python
sns.kdeplot(iris.Petal_Length[iris.Class=='Iris-setosa'],label='Setosa')
sns.kdeplot(iris.Petal_Length[iris.Class=='Iris-versicolor'],label='Versicolor')
sns.kdeplot(iris.Petal_Length[iris.Class=='Iris-virginica'],label='Virginica')
plt.legend()
plt.show
```

### Box Plot:- Used to display the statistical data of the variable.

```python
sns.boxplot(x=iris.Class,y=iris.Sepal_Width,palette='cool')
plt.show()
```

```python
sns.boxplot(x=iris.Class,y=iris.Sepal_Width,palette='hot')
plt.show()
```

### Violin Plot:- Used to display the distribution, median and density of the datas accross the categories.

```python
sns.violinplot(x=iris.Class,y=iris.Sepal_Width,palette='hot')
plt.show()
```

### Swarm Plot:- Used to display the statistical datas of each categories using all data points.

```python
sns.swarmplot(x=iris.Class,y=iris.Sepal_Length,palette='colorblind')
plt.show()
```

### Bar Plot:- Used to display the summary statistic(mean) of a variable along with a categorical data.
<pre>It automatically represents the Confidence interval(CI)
CI:- Range of values that are actually true</pre>

```python
sns.barplot(x=iris.Class,y=iris.Sepal_Width,palette='hot')
plt.show()
```

### Count Plot:- Shows the count of each categories.


```python
plt.figure(figsize=(20,15))
sns.countplot(x=iris.Sepal_Length,palette='cool')
plt.show()
```

```python
sns.countplot(x=iris.Class,data=iris,palette='hot')
plt.show()
```

### Scatter Plot:- Used to display relationship between two numerical values.

```python
sns.scatterplot(x=iris.Sepal_Length,y=iris.Sepal_Width,hue=iris.Class)
plt.show()
```

### Joint Plot:- Used to display the bivariate scatter plot of each variable.

```python
sns.jointplot(data=iris)
plt.show()
```

```python
sns.jointplot(data=iris,kind='kde')
plt.show()
```

### Pair Plot:- Creates a grid of plots that displays the pairwise relationship between multiple numerical columns in the dataset.

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='Class')
plt.show()
```
