#### SVM :- Support Vector Machine
#### SVC :- Support Vector Classifier :- to find the best fit hyperplane/line that separate the categories in the best fit way.

##### Kernel Function:- Aims to convert 1D to 2D and 2D to 3D
* Linear -> 1D to 2D
* Poly -> 2D to 3D
* Sigmoid -> 2D to 3D
* rbf (Radial Basis Function) -> 1D to 2D  & 2D to 3D

##### Gamma :- Gamma is known as the coefficient of the kernel function.
##### -> It is used for cancellation of noise

##### C :- This is known as Regularization parameter.
##### -> Restricts our model to be overfitted.
##### -> Range of C varies from 
#####                           (0.01 to 1) -> Small(noisy data, wants to avoid overfitting)
#####                           (1 to 10)   -> Medium(good training starting point)
#####                           (10 to 1000)-> High(Data is clean, helps to minimize overfitting)

```python
from sklearn.svm import SVC
sv = SVC(kernel = 'rbf', gamma = 0.01, C = 1000)
sv.fit(xtrain,ytrain)
```

```python
pred2 = sv.predict(xtest)
pred2
```

```python
from sklearn.metrics import recall_score,accuracy_score
rs = recall_score(pred2,ytest)
ac = accuracy_score(pred2,ytest)
print(f"Accuracy:",ac)
print(f"Recall:",rs)
```

```python
# Confusion Matrix
from sklearn.metrics import  ConfusionMatrixDisplay,confusion_matrix
cm = confusion_matrix(ytest,pred2)
cm2 = ConfusionMatrixDisplay(cm)
cm2.plot()
plt.show()
```
