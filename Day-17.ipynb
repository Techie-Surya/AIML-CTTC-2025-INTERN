#### Color Combination:-

```python
r = np.full((50,50),0)
g = np.full((50,50),0)
b = np.full((50,50),0)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1)
plt.axis('off')
plt.show()
```

```python
r = np.full((50,50),100)
g = np.full((50,50),60)
b = np.full((50,50),234)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1)
plt.axis('off')
plt.show()
```

```python
r = np.full((50,50),0)
g = np.full((50,50),255)
b = np.full((50,50),0)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1)
plt.axis('off')
plt.show()
```

```python
r = np.full((50,50),0)
g = np.full((50,50),255)
b = np.full((50,50),0)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1,alpha=0.5)
plt.axis('off')
plt.show()
```

### Creating Shapes:-

```python
# Drawing a line
# Syntax: cv2.line(startcoordinate(x,y),endcoordinate(x1,y1),thickness,color(RGB))
```

```python
# Creating an array of black colour
r = np.full((50,50),0)
g = np.full((50,50),0)
b = np.full((50,50),0)
rgb1 = np.dstack((r,g,b))
plt.imshow(rgb1)
plt.axis('off')
plt.show()
```

```python
p = np.zeros((500,500,3))
plt.imshow(p)
plt.show()
```

```python
cv2.line(p,(0,500),(500,0),(255,255,255),4)
plt.imshow(p)
plt.show()
```

```python
# cv2.circle(center(x,y)radius,color,thickness)
p1 = np.zeros((500,500,3))
plt.imshow(p1)
cv2.circle(p1,(250,250),50,(255,255,255),4)
plt.imshow(p1)
plt.show()
```

```python
cv2.circle(p1,(250,250),70,(255,255,255),4)
plt.imshow(p1)
plt.show()
```

```python
cv2.circle(p1,(250,250),90,(255,255,255),4)
plt.imshow(p1)
plt.show()
```

```python
cv2.circle(p1,(250,250),110,(255,255,255),4)
plt.imshow(p1)
plt.show()
```

```python
# cv2.rectangle(1stDiagonal(x,y),2ndDiagonal(x1,y1),color,thickness)
p2 = np.zeros((500,500,3))
cv2.rectangle(p2,(200,400),(300,200),(255,255,255),4)
plt.imshow(p2)
plt.show()
```

```python
# Converting BGR to Gray:-
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
cv2.imshow('Gray',gray)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
# Converting BGR to RGB:-
rbg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rbg)
plt.show()
```

```python
# Detection of Edges:-
edge = cv2.Canny(rbg,10,200)
plt.imshow(edge,cmap='gray')
plt.show()
```
